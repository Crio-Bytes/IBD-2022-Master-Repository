# Activity 2: Understanding the big picture

### Study of Relationship between Variables

The goal now is to uncover interesting correlations that demonstrate the influence of one variable on another, preferably the aim.

In a corporate setting such as digital marketing or online advertising, this information provides value and the opportunity to act strategically.

With the help of Seaborn and pairplot, we can begin studying relationships.

```python
sns.pairplot(df)
```

In a scatterplot, pairplot plots all variables against each other. It is quite helpful for grasping the main significant linkages without having to manually go through every single combination.

A boxplot is the greatest approach to comprehend the relationship between a numerical variable and a category variable.

Let's plot alcohol, flavanoids, color intensity, and proline in a boxplot. Why are these variables used? Because they reveal slightly more distinct segmentations for a specific wine type.

```python
sns.catplot(x="target", y="proline", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs proline")
plt.show()
```

We can see that type 0 wine has a higher proline median than the other two varieties. Is it a differentiating factor? It's too early to tell. There could be other factors to consider.

```python
sns.catplot(x="target", y="flavanoids", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs flavanoids")
plt.show()
```

The flavanoids content of type 0 wine appears to be higher. Is it feasible that type 0 wines have higher quantities of proline and flavanoids combined? We may use Seaborn to construct a scatterplot and see which wine class a point belongs to. Simply enter the hue parameter.

```python
sns.scatterplot(x="proline", y="flavanoids", hue="target", data=df, palette="Dark2", s=80)
plt.title("Relationship between proline, flavanoids and target")
plt.show()
```

Our intuition had been correct! Flavanoids and proline patterns are visible in Type 0 wines. Proline levels, in instance, are substantially greater, while flavanoid levels remain consistent around 3.

We will use Pandas to generate a correlation matrix and isolate the most associated variables.

```python

corrmat = df.corr()
fig, ax = plt.subplots(figsize=(10,10))         
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r",
                ax=ax)
plt.show()
```

The heat map is useful because it allows us to efficiently grasp which variables are strongly correlated with each other.

When the target variable decreases (which must be interpreted as a tendency to go to 0, therefore to the type of wine 0) the flavanoids, total phenols, proline and other proteins tend to increase. And viceversa.
